# 1. Sparse Representation
원-핫 벡터는 단어 유사도 표현 불가   

# 2. Distributed Representation
단어 의미를 다차원 공간에 벡터화      
희소 표현은 고차원에 단어를 차원별로 분리, 분산 표현은 저차원에 단어 의미를 차원별로 분산

# 3. Continuous BoW
**Word2Vec**는 입력이 원-핫 벡터인 CBoW와 Skip-gram   
은닉층 lookup table 하나인 Shallow Neural Network로 활성화 함수가 없음   

CBoW는 주변 단어로 중심 단어 예측   
![CBoW](https://wikidocs.net/images/page/22660/%EB%8B%A8%EC%96%B4.PNG "CBoW")

CBoW 인공 신경망   
![CBoW](https://wikidocs.net/images/page/22660/word2vec_renew_1.PNG "CBoW")

lookup table은 가중치 행렬 i번째 행을 그대로 읽음   
lookup table 결과 벡터들은 투사층에서 평균 벡터가 됨   
평균 벡터 두번째 가중치 행렬 곱에 softmax를 취해 스코어 벡터 계산   
![CBoW](https://wikidocs.net/images/page/22660/word2vec_renew_3.PNG "CBoW")

# 4. Skip-gram
Skip-gram은 중심 단어로 주변 단어 예측
![Skip-gram](attachment:image.png "Skip-gram")

Skip-gram 인공 신경망   
투사층에서 평균 벡터를 구하지 않음   
![Skip-gram](https://wikidocs.net/images/page/22660/word2vec_renew_6.PNG "Skip-gram")

# 5. NNLM vs. Word2Vec
NNLM: 언어 모델로 다음 단어 예측시 앞 단어만 참고   
Word2Vec: 워드 임베딩으로 중심 단어 예측시 앞뒤 모든 단어 참고   
![NNLM vs. Word2Vec](https://wikidocs.net/images/page/22660/word2vec_renew_7.PNG "NNLM vs. Word2Vec")
