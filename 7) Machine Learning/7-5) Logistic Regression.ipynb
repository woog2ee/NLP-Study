# 2. 시그모이드 함수(Sigmoid Function)
$$H(X) = \frac{1}{1+e^{-(Wx+b)}} = sigmoid(Wx + b) = \sigma(Wx + b)$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0], [1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

```python
# w 값에 따른 sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b', linestyle='--')
plt.plot([0,0], [1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

```python
# b 값에 따른 sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--')
plt.plot(x, y2, 'g')
plt.plot(x, y3, 'b', linestyle='--')
plt.plot([0,0], [1.0,0.0], ':')
plt.title('Sigmoid Function')
plt.show()
```

# 3. 비용 함수(Cost Function)
가중치를 최소로 하는 비용 함수이자 목적 함수(objective function)   
로지스틱 회귀는 경사 하강법으로 가중치를 찾으나 비용 함수로 MSE를 사용하진 않음   
global minimum이 아닌 local minimum을 최솟값으로 도출할 가능성 때문   
$$J(W) = \frac{1}{n}\sum_{i=1}^{n}cost(H(x^{(i)}), y^{(i)})$$

실제값이 1일 때 y가 0일수록 오차가 커짐   
실제값이 0일 때 y가 1일수록 오차가 커짐   
$$if\, y = 1,\, cost(H(x), y) = -log(H(x))$$   
$$if\, y = 0,\, cost(H(x), y) = -log(1 - H(x))$$

```python
x = np.arange(0, 1, 0.01)
y_1 = -np.log(x)
y_0 = -np.log(1-x)

plt.plot(x, y_1, 'b')
plt.plot(x, y_0, 'r')
plt.title('Cost Function')
plt.show()
```

로지스틱 회귀 비용 함수는 크로스 엔트로피(cross entropy) 함수   
가중치를 찾고자 크로스 엔트로피 함수의 평균을 취해 사용
$$cost(H(x), y) = -[ylogH(x) + (1 - y)log(1 - H(x))]$$   
$$J(W) = -\frac{1}{n}\sum_{i=1}^{n}[y^{(i)}logH(x^{(i)}) + (1 - y^{(i)})log(1 - H(x^{(i)}))]$$
