# 1. 다중 선형 회귀
$$H(X) = W_{1}x_{1} + W_{2}x_{2} + ... + b$$

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

X = np.array([[70,85,11],[71,89,18],[50,80,20],[99,20,10],[50,10,10]]) # 중간, 기말, 가산점
y = np.array([73,82,72,57,34]) # 최종 성적
# 입력 벡터 차원=3, 출력 벡터 차원=1

model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))

sgd = optimizers.SGD(lr=0.00001)
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])

model.fit(X, y, batch_size=1, epochs=200, shuffle=False)
```

```python
print(model.predict(X))
```

```python
X_test = np.array([[20,99,10],[40,50,20]])
print(model.predict(X_test))
```

# 2. 다중 로지스틱 회귀
$$H(X) = sigmoid(W_{1}x_{1} + W_{2}x_{2} + ... + b)$$

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,1])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])

model.fit(X, y, batch_size=1, epochs=800, shuffle=False)
```

```python
print(model.predict(X))
```

# 3. 인공 신경망 다이어그램
다중 로지스틱 회귀를 일종의 인공 신경망 구조로 해석 가능   
$$y = sigmoid(W_{1}x_{1} + W_{2}x_{2} + ... + b) = \sigma(W_{1}x_{1} + W_{2}x_{2} + ... + b)$$
